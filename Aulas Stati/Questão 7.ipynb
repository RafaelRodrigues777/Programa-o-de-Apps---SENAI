Questão 7: Monitoramento de Purificação de Água (Tratamento de Água)
O setor de tratamento de água coleta amostras para monitorar a concentração de impurezas (em
ppm). Os dados são registrados diariamente, mas o sistema requer limpeza, remoção de valores
extremos e organização para relatórios.
1. Crie uma lista impurezas_ppm com os seguintes dados de medição: [12.5, 15.0, 11.2, 12.5,
18.0, 14.8, 11.2, 20.1].

```python
impurezas_ppm = [12.5, 15.0, 11.2, 12.5,
18.0, 14.8, 11.2, 20.1]
print(impurezas_ppm)
```

2. Os dados de uma segunda estação de tratamento precisam ser adicionados. Crie uma lista
segunda_estacao = [13.5, 16.2, 14.0]. Use extend() para adicionar esses valores a
impurezas_ppm.

```python
segunda_estacao = [13.5, 16.2, 14.0]
print(segunda_estacao)

impurezas_ppm.extend(segunda_estacao)
print(impurezas_ppm)
```

3. Identifique o valor 11.2 na lista e use remove() para remover apenas a primeira ocorrência
dele.

```python
for i in impurezas_ppm:
    if i == 11.2:
        impurezas_ppm.remove(11.2)
        break
print(impurezas_ppm)
```

4. Use um laço for para percorrer a lista impurezas_ppm e conte quantas vezes a concentração
12.5 foi registrada. Imprima o resultado.

```python
total = 0

for i in impurezas_ppm:
    if i == 12.5:
        total += 1  
    
print(total)
```

5. O valor mais alto de impureza (o último item na lista após a ordenação crescente) deve ser
removido para evitar distorções na média. Use sort() para ordenar a lista e, em seguida, use
pop() para remover o valor mais alto.

```python
impurezas_ppm.sort()
print(impurezas_ppm)

impurezas_ppm.pop()
print(impurezas_ppm)
```

6. Um novo registro 10.9 de uma amostra de alta qualidade foi obtido e precisa ser inserido na
lista, mantendo a ordenação. Use insert() para adicioná-lo na posição correta.


```python
impurezas_ppm.insert(0, 10.9)
print(impurezas_ppm)
```

7. Imprima a lista final de impurezas.

```python
print(impurezas_ppm)
```
