Questão 8: Gerenciamento de Dados de Colaboradores (Recursos Humanos)
O setor de RH da Bosch precisa organizar e atualizar uma lista de colaboradores para um projeto
multissetorial.
1. Crie uma lista chamada colaboradores_bosch com os seguintes nomes de colaboradores:
['Ana Silva', 'Bruno Costa', 'Carlos Mendes', 'Diana Pires', 'Ana Silva', 'Eduardo Santos',
'Fátima Rocha', 'Ana Silva'].

```python
colaboradores_bosch = ['Ana Silva', 'Bruno Costa', 'Carlos Mendes', 'Diana Pires', 'Ana Silva', 'Eduardo Santos',
'Fátima Rocha', 'Ana Silva']
print(colaboradores_bosch)
```

2. Use o método count() para verificar quantas vezes o nome 'Ana Silva' aparece na lista e
imprima o resultado.

```python
total = colaboradores_bosch.count('Ana Silva')
print(total)
```

3. Um novo colaborador, 'Gabriel Souza', foi contratado para a área de Análise de Dados. Use
append() para adicioná-lo ao final da lista.

```python
colaboradores_bosch.append('Gabriel Souza')
print(colaboradores_bosch)
```

4. O colaborador 'Carlos Mendes' foi transferido para outra unidade e precisa ser removido da
lista. Use remove() para retirá-lo.

```python
colaboradores_bosch.remove('Carlos Mendes')
print(colaboradores_bosch)
```

5. Há um novo colaborador, 'Isabel Torres', que se juntou à equipe e deve ser inserido na
terceira posição da lista (índice 2) para manter a ordem de entrada no projeto. Use insert()
para adicioná-la.

```python
colaboradores_bosch.insert(2, 'Isabel Torres')
print(colaboradores_bosch)
```

6. A lista de colaboradores precisa ser ordenada em ordem alfabética para facilitar a busca.
Use sort() para ordenar a lista.

```python
colaboradores_bosch.sort()
print(colaboradores_bosch)
```

7. Por fim, o colaborador 'Bruno Costa' infelizmente teve que sair do projeto e deve ser
removido. Use pop() para remover este item. Se precisar, use index() para encontrar a
posição dele antes. Imprima a lista final.

```python
indice = colaboradores_bosch.index('Bruno Costa')
print(indice)

colaboradores_bosch.pop(3)
print(colaboradores_bosch)
```
