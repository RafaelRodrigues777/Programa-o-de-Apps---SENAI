Questão 6: Análise de Qualidade de Brocas (Setor de Brocas)
A equipe de qualidade registrou os diâmetros de um lote de brocas. No entanto, o sistema de
medição gerou alguns dados duplicados e registros que precisam ser corrigidos e organizados.
1. Crie uma lista chamada diametros_brocas com os seguintes valores: [5.0, 5.2, 5.1, 5.0, 5.3,
5.0, 5.4].

```python
diametros_brocas =[5.0, 5.2, 5.1, 5.0, 5.3,
5.0, 5.4]

print(diametros_brocas)
```

2. Um novo lote de brocas foi medido. Crie uma nova lista novos_diametros = [5.5, 5.1, 5.6].
Use extend() para adicionar os novos valores à lista diametros_brocas.

```python
novos_diametros = [5.5, 5.1, 5.6]
print(novos_diametros)

diametros_brocas.extend(novos_diametros)
print(diametros_brocas)
```

3. Utilize um laço for para iterar sobre a lista diametros_brocas. Dentro do laço, use count()
para verificar a quantidade de vezes que o valor 5.0 aparece. Em seguida, use remove() para
remover todas as ocorrências de 5.0 da lista.

```python
total = 0
total += diametros_brocas.count(5.0)

for i in diametros_brocas:
    if i == 5.0:
        diametros_brocas.remove(5.0)
print(total)
print(diametros_brocas)
```

4. O registro de um diâmetro de broca crítico (4.9) precisa ser adicionado. Use insert() para
colocar este valor na segunda posição da lista (índice 1).

```python
diametros_brocas.insert(1, 4.9)
print(diametros_brocas)
```

6. O diâmetro da broca com o maior valor na lista (o primeiro item após a ordenação
decrescente) precisa ser removido. Use pop() para fazer isso e imprima o valor removido.

```python
diametros_brocas.sort()
print(diametros_brocas)

diametros_brocas.pop()
print(diametros_brocas)
```

7. Imprima a lista final após todas as operações.

```python
print(diametros_brocas)
```
