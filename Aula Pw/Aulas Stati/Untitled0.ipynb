```python
import pandas as pd

# Para ler arquivos Excel, você talvez precise instalar uma biblioteca extra.
# Se receber um erro, abra o terminal e rode: pip install openpyxl

# Carrega a planilha do Excel para um DataFrame
df = pd.read_excel('dados_cafeteria.xlsx')

# --- Testes e Análises Básicas ---

# 1. Veja as 5 primeiras linhas para entender os dados
print("--- Primeiras 5 linhas dos dados ---")
print(df.head())
print("\n" + "="*50 + "\n")

# 2. Veja informações gerais (tipos de colunas, contagem de dados)
print("--- Informações gerais do DataFrame ---")
df.info()
print("\n" + "="*50 + "\n")

# 3. Calcule uma nova coluna com o total de cada venda
df['Total_Venda'] = df['Preco_Unitario'] * df['Quantidade']
print("--- Dados com a nova coluna 'Total_Venda' ---")
print(df.head())
print("\n" + "="*50 + "\n")

# 4. Calcule o faturamento total da cafeteria
faturamento_total = df['Total_Venda'].sum()
print(f"Faturamento Total: R$ {faturamento_total:.2f}")
print("\n" + "="*50 + "\n")

# 5. Agrupe para ver o total de vendas por categoria de produto
vendas_por_categoria = df.groupby('Categoria')['Total_Venda'].sum()
print("--- Vendas Totais por Categoria ---")
print(vendas_por_categoria)
```
