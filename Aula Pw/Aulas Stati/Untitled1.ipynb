```python
import pandas as pd

data = {
    'Nome': ['Ana', 'Bruno', 'Ana', 'Diana'],
    'Idade': [23, 35, 23, 40]
}
df = pd.DataFrame(data)

# Estatísticas da coluna Idade (média, min, max, etc)
print("Descrição Idade:")
print(df['Idade'].describe())

# Estatísticas da coluna Nome (contagem, valores únicos, valor mais frequente)
print("\nDescrição Nome:")
print(df['Nome'].describe())

# Contar quantas vezes cada nome aparece
print("\nContagem dos nomes:")
print(df['Nome'].value_counts())

# Mostrar os nomes únicos na coluna Nome
print("\nNomes únicos:")
print(df['Nome'].unique())

# Filtrar linhas onde Idade é maior que 30
print("\nPessoas com Idade > 30:")
print(df[df['Idade'] > 30])

# Adicionar uma nova coluna "Idade em 5 anos"
df['Idade + 5 anos'] = df['Idade'] + 5
print("\nDataFrame com nova coluna Idade + 5 anos:")
print(df)

# Ordenar DataFrame pela coluna Idade (do menor para maior)
print("\nDataFrame ordenado pela Idade:")
print(df.sort_values(by='Idade'))

# Resetar índice após ordenação
print("\nDataFrame ordenado pela Idade com índice resetado:")
print(df.sort_values(by='Idade').reset_index(drop=True))
```
