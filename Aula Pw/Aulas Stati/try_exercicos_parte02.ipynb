Exercício 1. Faça um programa que receba uma idade, e realize as seguintes verificações:
* Se a pessoa tiver menos de 16 anos, a função deve lançar um erro com a mensagem: "Você 
não tem idade para votar.";
* Se a pessoa tiver entre 16 e 17, ou mais que 70 anos, a função deve imprimir a mensagem: 
"Você pode votar, mas o voto é facultativo.";
* Se a pessoa tiver entre 18 e 70 anos, a função deve imprimir a mensagem: "Você deve votar, é 
obrigatório.";
* Além disso, deve possuir verificações para valores negativos ou inválidos.

Exercício 2. Implemente um programa que valide uma senha, ele deve possuir os seguintes critérios:
* Ter pelo menos 8 caracteres;
* Conter pelo menos uma letra maiúscula, uma letra minúscula e um número;
* Conter pelo menos um caractere especial (como @, #, $, etc.);

Exercício 3. Implemente um programa que receba a data de hoje e uma data de nascimento e realize as 
seguintes validações:
* Verificar se as datas foram enviadas no formato correto (dd/mm/aaaa);
*  Validar se a quantidade de dias e meses é correta para o formato;
* Verificar se o ano informado é válido, ou seja, o ano de nascimento não pode ser, por exemplo, 
2030;
* Calcular a idade da pessoa considerando a data de nascimento e a data de hoje informada;
*  Validar se a idade calculada está coerente, ou seja, deve ser maior que 0 e menor que 100 anos.
O programa deve retornar uma mensagem de erro se algum dos critérios de validação falhar. Caso as datas 
estejam corretas, o programa deve exibir a idade da pessoa.

Exercício 4. Implemente um programa que peça ao usuário para digitar o nome do produto, o preço original e 
o percentual de desconto. Ele deve validar:
* O nome do produto, que deve ser escrito sem números ou caracteres especiais;
* O preço, que deve ser um número válido e positivo;
* O percentual de desconto, que deve ser um número entre 0 e 100.
Exiba o valor inicial (preço original) e o valor final (com desconto) formatados como moeda, além do percentual 
de desconto com duas casas decimais.
