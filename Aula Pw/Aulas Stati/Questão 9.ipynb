Qustão 9: Gestão de Lotes de Produção e Análise de Defeitos (Setor de Produção)
A equipe de produção da Bosch registrou os códigos de lote de produtos fabricados, mas precisa
realizar uma análise complexa para identificar e gerenciar lotes com possíveis defeitos e organizar o
registro final.
1. Crie uma lista chamada lotes_producao com os seguintes códigos: ['A123', 'B456', 'C789',
'D012', 'A123', 'E345', 'F678', 'A123', 'G901'].

```python
lotes_producao = ['A123', 'B456', 'C789',
'D012', 'A123', 'E345', 'F678', 'A123', 'G901']
print(lotes_producao)
```

2. Um novo turno de produção gerou novos lotes. Crie uma lista novos_lotes = ['H234', 'I567'] e
use extend() para adicionar esses códigos à lista lotes_producao.

```python
novos_lotes = ['H234', 'I567']
lotes_producao.extend(novos_lotes)

print(lotes_producao)
```

3. Utilize um laço for para iterar sobre a lista lotes_producao e, dentro do laço, conte quantas
vezes o código 'A123' aparece. Após a contagem, use remove() para remover todas as
ocorrências de 'A123' da lista.

```python
total = 0
total += lotes_producao.count('A123')

for i in lotes_producao:
    if i == 'A123':
        lotes_producao.remove('A123')
print(total)
print(lotes_producao)
```

4. O código de um lote que apresentou defeitos ('D012') deve ser removido, pois foi
descartado. Use pop() para remover este código e imprima o valor removido.

```python
removido = lotes_producao.pop(2)
print(removido)
```

5. O setor de qualidade identificou que um novo lote com código 'Z999' precisa ser inserido na
segunda posição da lista (índice 1) para priorizar a análise. Use insert() para adicionar este
novo código.

```python
lotes_producao.insert(1,'Z999')
print(lotes_producao)
```

6. A lista de lotes precisa ser ordenada alfabeticamente para um relatório. Use sort() para
ordenar a lista.

```python
lotes_producao.sort()
print(lotes_producao)
```

7. Por fim, adicione um novo lote chamado 'X888' ao final da lista, usando append().

```python
lotes_producao.append('X888')
print(lotes_producao)
```

8. Imprima a lista final de lotes de produção após todas as operações.

```python
print(lotes_producao)
```
