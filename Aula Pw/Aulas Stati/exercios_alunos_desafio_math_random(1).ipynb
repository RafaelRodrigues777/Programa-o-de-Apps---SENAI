### Usando biblioteca Math

_Consulte a documentação das bibliotecas_

1. Um círculo tem um raio de r metros. Qual expressão Python usa a biblioteca `math` para calcular a área desse círculo ( Area=πr²)? - uso metodo para pi e potencia e resultado com 2 casas decimais. 

```python
import math as m

raio = float(input("Insira o raio"))

area = m.pi*m.pow(raio,2)

print(F"{area:.2f}")
```

2. Calcule o Perímetro (Circunferência) do círculo (Perimetro=2πr) usando metodo pi e usuário insere o valor do raio. Imprima o resultado com duas casas decimais.

```python
import math as m

raio = float(input("Insira o raio"))


perimetro = 2*m.pi*raio

print(F"{perimetro:.2f}")
```

3. Em um triângulo retângulo, os catetos medem 3 e 4 unidades. Qual função da biblioteca `math` permite calcular a hipotenusa de forma mais eficiente (Teorema de Pitágoras, h² =a² +b²)?- usar a biblioteca `hypo`

```python
import math as m

cateto1 = 3
cateto2 = 4

hypo = m.hypot(cateto1,cateto2)
print(f"{hypo:.2f}")
```

4. Calcule a raiz quadrada e fatorial usando `math`. Crie um programa que faça o seguinte:

* Defina uma variável `numero_base`.

* Calcule a raiz quadrada de `numero_base`.

* Calcule o fatorial da raiz quadrada (o fatorial do resultado do passo anterior) usando `factorial`.

* Imprima a __raiz quadrada__ e o resultado do __fatorial__.

```python
numero_base = int(input("Declare o numero base"))

raiz = m.sqrt(numero_base)
raiz = int(raiz)
fatorial = m.factorial(raiz)

print(f"A raiz de {numero_base} é {raiz} e o fatorial é {fatorial}")
```

 Se você tem uma lista de pessoas (João, Fabio, Eduardo, Enzo, Gabriela, Padilha, Janice), qual função você usaria para sortear aleatoriamente **um** item desta lista?

```python
import random as r

listaNomes = ["João", "Fabio", "Eduardo", "Enzo", "Gabriela", "Padilha", "Janice"]

print(r.choice(listaNomes))
```

Com a mesma lista anterior, sortear:
* um resultado com duas pessoas, mas pode ser um resultado com dois nomes iguais
* outro resultado com duas pessoas, porém não pode repetir

```python
import random as r

meus_itens = ["João", "Fabio", "Eduardo", "Enzo", "Gabriela", "Padilha", "Janice"]

for i in range(0,2):
    print(r.choice(listaNomes))
numero_de_selecoes = 2

itens_aleatorios = r.sample(meus_itens, numero_de_selecoes)

print(itens_aleatorios)
```
