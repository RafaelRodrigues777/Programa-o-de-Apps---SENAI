1. Você precisa criar uma calculadora de descontos. Desenvolva uma função chamada `aplicar_desconto` que receba três argumentos: o `preco_original` de um produto, a `porcentagem_desconto` (como um número de 0 a 100, por exemplo, 10 para 10%) e um `tipo_desconto` (uma string que pode ser 'percentual' ou 'fixo').

- Se `tipo_desconto` for 'percentual', o desconto deve ser calculado com base na porcentagem.
- Se `tipo_desconto` for 'fixo', o desconto deve ser subtraído diretamente do preço original.

A função deve **retornar** o `preco_final` após a aplicação do desconto.

Considere os seguintes casos:
- Se `porcentagem_desconto` for inválida (menor que 0 ou maior que 100) quando o tipo for 'percentual', retorne uma mensagem de erro como string.
- Se o `preco_original` for menor ou igual a zero, retorne uma mensagem de erro.

2. Crie um programa que declare uma ```lista``` chamada idades com 5 números inteiros. Logo em seguida, crie uma função chamada ```adicionar``` que receba uma lista como parâmetro e que *adicione* um novo elemento qualquer a essa lista. Feito isso, chame a função adicionar passando a lista idades como argumento. Por fim, após a chamada da função, mostre a lista idades para verificar se o elemento foi adicionado corretamente.
