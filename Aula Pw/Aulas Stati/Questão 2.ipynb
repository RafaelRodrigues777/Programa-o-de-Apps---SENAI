Questão 2: Análise de Desempenho de Linha de Produção
A equipe de produção registrou as horas de parada de uma linha de montagem durante um mês. A
lista paradas_diarias representa as horas de inatividade. O objetivo é remover os registros de
menor impacto e ordenar os restantes para análise.
1. Crie uma lista paradas_diarias com os seguintes valores: [2.5, 1.0, 0.5, 3.2, 1.5, 0.5, 4.0, 2.1].

```python
paradas_diarias = [2.5, 1.0, 0.5, 3.2, 1.5, 0.5, 4.0, 2.1]
```

2. Use sort() para ordenar a lista em ordem crescente.

```python
paradas_diarias.sort()
print(paradas_diarias)
```

3. Identifique o valor que representa a menor parada na lista (o primeiro item após a
ordenação).

```python
indice = paradas_diarias.index(1)
print(indice)
```

Use pop() para remover o último item da lista. Imprima o item removido e a lista atualizada.

```python
paradas_diarias.pop()
print(paradas_diarias)
```

Ocorreu um registro de parada importante que deve ser inserido em uma posição específica.
Use insert() para adicionar o valor 3.5 na posição de índice 3.

```python
paradas_diarias.insert(3,3.5)
print(paradas_diarias)
```

Imprima a lista final após todas as operações.

```python
print(paradas_diarias)
```
