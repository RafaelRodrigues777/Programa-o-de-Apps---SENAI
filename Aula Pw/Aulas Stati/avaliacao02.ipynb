### Tarefa 1: Sistema de Priorização de Manutenção Preditiva (```match-case```)

_Cenário_: A Bosch utiliza um sistema automatizado para priorizar chamados de manutenção com base no código de severidade reportado pelos sensores.

_Instruções_:
Crie um script Python que receba um ```codigo_severidade``` (string) como entrada (usar recurso para deixar tudo em caixa-alta se o usuário digitar em caixa baixa) e determine a ação de prioridade. Use a instrução ```match-case``` para implementar a seguinte lógica:

* Se ```codigo_severidade``` for ```"CRITICO"```, imprima: "Prioridade MÁXIMA: Intervenção imediata necessária para evitar falha total."

* Se ```codigo_severidade``` for ```"ALERTA_ALTO"```, imprima: "Prioridade ALTA: Agendar manutenção em até 24 horas. Risco de impacto na produção."

* Se ```codigo_severidade``` for ```"AVISO"```, imprima: "Prioridade MÉDIA: Agendar manutenção preventiva para a próxima janela disponível. Risco baixo."

Para qualquer outro ```codigo_severidade```, imprima: "Status: Normal. Nenhuma ação de manutenção prioritária necessária."

Entrada: "CRITICO"
Saída: Prioridade MÁXIMA: Intervenção imediata necessária para evitar falha total.

Entrada: "INFO"
Saída: Status: Normal. Nenhuma ação de manutenção prioritária necessária.

###  Tarefa 2: Classificação de Qualidade de Componentes Eletrônicos (```match-case``` com Múltiplos Padrões)

_Cenário_: Componentes eletrônicos recebidos de fornecedores são classificados por um sistema de controle de qualidade baseado em um código numérico.

_Instruções_:
Desenvolva um script Python que recebe um ``````codigo_qualidade`````` (inteiro) e imprime uma mensagem descrevendo o status do lote. Utilize ```match-case``` com múltiplos padrões:

Entrada: 4
Saída: Lote Rejeitado: Componentes com defeitos significativos. Encaminhar para descarte ou reprocessamento.

Entrada: 2
Saída: Lote Aprovado: Componentes de alta qualidade. Liberado para produção.

### Tarefa 3: Monitoramento de Consumo de Energia (```while``` com ```if```)

_Cenário_: Uma máquina na fábrica Bosch consome energia. O sistema monitora o consumo em intervalos regulares e emite um alerta se o consumo exceder um limite crítico. A monitorização continua até que o alerta seja emitido ou um número máximo de verificações seja atingido.

_Instruções_:
* Crie um script que simule o monitoramento de consumo de energia.

* Inicialize ``````consumo_atual`````` em ```1500``` Watts e limite_critico em  ```2000``` Watts.

* Inicialize ```verificacoes_realizadas``` em ```0``` e um ```max_verificacoes``` em ```10```.

* Use um loop while que continue enquanto ```verificacoes_realizadas < max_verificacoes``` E ```consumo_atual < limite_critico```.

Dentro do loop:

* Simule um aumento aleatório no ```consumo_atual``` (você pode adicionar um valor fixo ou usar ```random.randint``` da library ```random```). Por exemplo, adicione entre 50 e 150 Watts.

* Imprima o ```consumo_atual``` em cada passo, indicando que é uma verificação.

* Incrementa verificacoes_realizadas.

Após o loop, verifique se o ```consumo_atual``` atingiu ou ultrapassou o ```limite_critico```.

* Se sim, imprima: "ALERTA DE CONSUMO CRÍTICO! Desligar a máquina para verificação."

* Se não (ou seja, o loop terminou por atingir ```max_verificacoes```), imprima: "Consumo de energia dentro dos limites aceitáveis após 10 verificações."

### Tarefa 4: Processamento em Lotes de Peças Usando ```for``` e Verificação de Qualidade

_Cenário_: Um lote de 10 peças foi produzido. Cada peça tem um código de lote. É preciso verificar se alguma peça contém um código de lote "defeituoso".

_Instruções_:
Crie um script Python.

* Crie uma lista ```codigos_pecas``` contendo 10 códigos de lote. Inclua o código ```"ERRO_LOTE_XYZ" ```em alguma posição da lista (por exemplo, na 7ª posição).

* Use um loop for para iterar sobre a lista ```codigos_pecas```.

* Dentro do loop, use uma condicional ```if``` para verificar se o ```codigo_da_peca``` atual é igual a ```"ERRO_LOTE_XYZ"```.

* Se for encontrado o código de erro, imprima: "ERRO DETETADO: Peça com código 'ERRO_LOTE_XYZ' encontrada. Processo de inspeção interrompido." e utilize ```break``` para sair do loop.

* Caso contrário, imprima: "Verificando peça: ```codigo_da_peca```. OK."

Após o loop, use uma flag (variável booleana) para indicar se o erro foi encontrado. Se o loop terminar sem encontrar o erro, imprima: "Lote de peças verificado. Nenhum erro de lote crítico detectado."

### Tarefa 6: Processamento de Ordens de Serviço de Equipamentos (```while``` e ```for```)

_Cenário_: Várias ordens de serviço de manutenção para diferentes equipamentos da Bosch chegam ao sistema. Cada ordem tem um ID e uma lista de tarefas. Precisamos processar essas ordens sequencialmente.

_Instruções:_

1. Crie uma lista de dicionários, onde cada dicionário representa uma ```ordem_servico```. Cada ordem deve ter:

* ```id_ordem``` (ex: 1001, 1002)
* equipamento (ex: "Linha de Montagem A", "Máquina de Solda B")

* tarefas (uma lista de strings, ex: ```['Verificar rolamentos', 'Lubrificar engrenagens']```)

2. Crie uma variável ```ordens_a_processar``` inicializada com o número de ordens que você criou (por exemplo, 2).

3. Use um loop ```while``` que continue enquanto ```ordens_a_processar > 0```.

4. Dentro do ```while```:

* Obtenha a próxima ordem de serviço da sua lista (você pode usar o índice ou ```pop()``` para remover e processar).

* Imprima o ID da ordem e o nome do equipamento.

* Use um loop ```for``` para iterar sobre a lista de tarefas da ordem atual e imprima cada tarefa.

* Após processar todas as tarefas de uma ordem, imprima uma mensagem de ```"Ordem de Serviço #id_ordem concluída."```

* Decremente ```ordens_a_processar```.

5. Quando o loop ```while``` terminar, imprima: ```"Todas as ordens de serviço foram processadas."```
