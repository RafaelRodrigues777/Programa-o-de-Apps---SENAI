Questão 3: Consolidação de Dados de Sensores
Dois sensores de temperatura de uma máquina estão gerando dados em listas separadas. Você
precisa consolidar esses dados, limpá-los e prepará-los para visualização.
1. Crie duas listas: sensor_a = [23.4, 25.1, 22.9, 23.4, 26.0] e sensor_b = [24.5, 25.2, 23.8, 26.1].

```python
sensor_a = [23.4, 25.1, 22.9, 23.4, 26.0]
sensor_b = [24.5, 25.2, 23.8, 26.1]
```

2. Use extend() para combinar os dados do sensor_b ao final da lista sensor_a.

```python
sensor_a.extend([24.5, 25.2, 23.8, 26.1])
print(sensor_a)
```

3. O valor 23.4 foi registrado duas vezes em sensor_a. Use remove() duas vezes para remover
ambas as ocorrências. (Dica: remova uma de cada vez).

```python
sensor_a.remove(23.4)
print(sensor_a)

#sensor_a.remove(23.4)
#print(sensor_a)
```

4. Use sort() para ordenar a lista final de temperaturas em ordem decrescente.

```python
sensor_a.sort(reverse=True)
print(sensor_a)
```

5. Um dado de temperatura crítica (27.5) precisa ser adicionado à lista, mantendo a ordem.
Use insert() para adicioná-lo na posição correta.

```python
sensor_a.insert(0, 27.5)
print(sensor_a)
```
